# PyAAindex Example Usage

This notebook demonstrates how to use the `get_features` and `to_frame` functions from the `pyaaindex` package.

```python
import sys
sys.path.insert(0, './src') # Enables local import from the src directory

from pyaaindex.api import get_features, to_frame
```

## 1. Fetching Features (`get_features`)
We can query both single-index (AAindex1) and pair-index (AAindex2/AAindex3) features simultaneously.

```python
feature_ids = ["ARGP820101", "ALTS910101"]

# get_features will output alignment order for pair data in the console automatically
data = get_features(feature_ids)
```

### idx1 (AAindex1 DataFrame)
For `idx1`, the result is a pandas DataFrame. Rows are Canonical Amino Acids and Columns are the Feature IDs.

```python
df_idx1 = data['idx1']
display(df_idx1.head())
```

### idx2 & idx3 (AAindex2/AAindex3 Dictionaries)
For `idx2` and `idx3`, the result is a dictionary in the format `{feature_name: {aa1: [values]}}`. 
Pair values are alphabetically sorted arrays where the key is the first amino acid (`aa1`).

```python
dict_idx2 = data['idx2']

# Look at the values for 'A' against other amino acids
print("Values for AA1='A':")
print(dict_idx2['ALTS910101']['A'])
```

## 2. Converting Pair JSON to DataFrames (`to_frame`)
We can pass the `idx2` or `idx3` dictionaries directly to the `to_frame()` utility to get them as Pandas DataFrames with proper column mappings.

```python
frames = to_frame(data['idx2'])

pair_df = frames['ALTS910101']
display(pair_df.head())
```
